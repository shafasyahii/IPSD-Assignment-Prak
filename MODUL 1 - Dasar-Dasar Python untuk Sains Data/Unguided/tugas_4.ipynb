### Soal 4: Studi Kasus Pengelolaan Data
**Soal**: Anda diberikan file CSV berisi data nilai ujian mahasiswa. Tugas Anda adalah menulis sebuah program yang:
1. Membaca file CSV dan menyimpan datanya ke dalam dictionary.
2. Menghitung rata-rata nilai tiap mahasiswa.
3. Menampilkan mahasiswa dengan nilai tertinggi dan terendah.

**Tujuan**: Menguji kemampuan mahasiswa dalam membaca file, mengelola data dengan dictionary, serta menerapkan logika perhitungan.

---


```python
from google.colab import files
import pandas as pd
import io

# Mengunggah file CSV dari komputer Anda
upload_files = files.upload()

# Membaca file CSV yang diunggah
for filename in upload_files.keys():
    data = pd.read_csv(io.StringIO(upload_files[filename].decode('utf-8')))

# Menampilkan lima baris pertama dari data
print(data.head())

# Menampilkan ukuran data (jumlah baris dan kolom)
print("Ukuran Data:", data.shape)
```

```python
import csv

def baca_nilai_csv(nama_file):
    ## Membaca file CSV dan mengembalikan dictionary dengan nama mahasiswa sebagai kunci dan nilai sebagai list
    data_mahasiswa = {}

    with open(nama_file, mode='r') as file:
        reader = csv.reader(file)
        header = next(reader)  # Mengabaikan header

        for row in reader:
            nama = row[0]
            nilai = list(map(int, row[1:]))  # Mengonversi nilai menjadi integer
            data_mahasiswa[nama] = nilai

    return data_mahasiswa

def hitung_rata_rata(data_mahasiswa):
    ## Menghitung rata-rata nilai untuk setiap mahasiswa
    rata_rata = {}

    for nama, nilai in data_mahasiswa.items():
        if len(nilai) > 0:  # Pastikan list nilai tidak kosong
            rata_rata[nama] = sum(nilai) / len(nilai)
        else:
            rata_rata[nama] = 0  # Jika tidak ada nilai, rata-rata diatur ke 0

    return rata_rata

def mahasiswa_tertinggi_dan_terendah(rata_rata):
    ## Menentukan mahasiswa dengan nilai tertinggi dan terendah
    if not rata_rata:
        return None, None, None, None  # Jika tidak ada data

    mahasiswa_tertinggi = max(rata_rata, key=rata_rata.get)
    mahasiswa_terendah = min(rata_rata, key=rata_rata.get)

    return mahasiswa_tertinggi, rata_rata[mahasiswa_tertinggi], mahasiswa_terendah, rata_rata[mahasiswa_terendah]

# Nama file CSV
nama_file = 'nilai.csv'

# Membaca data dari file CSV
data_mahasiswa = baca_nilai_csv(nama_file)

# Menghitung rata-rata nilai tiap mahasiswa
rata_rata_nilai = hitung_rata_rata(data_mahasiswa)

# Menampilkan hasil
print("Rata-rata nilai tiap mahasiswa:")
for nama, rata in rata_rata_nilai.items():
    print(f"{nama}: {rata:.2f}")

# Menentukan dan menampilkan mahasiswa dengan nilai tertinggi dan terendah
mahasiswa_tertinggi, nilai_tertinggi, mahasiswa_terendah, nilai_terendah = mahasiswa_tertinggi_dan_terendah(rata_rata_nilai)

if mahasiswa_tertinggi and mahasiswa_terendah:
    print(f"\nMahasiswa dengan nilai tertinggi: {mahasiswa_tertinggi} ({nilai_tertinggi:.2f})")
    print(f"Mahasiswa dengan nilai terendah: {mahasiswa_terendah} ({nilai_terendah:.2f})")
else:
    print("\nTidak ada data untuk ditampilkan.")
```
